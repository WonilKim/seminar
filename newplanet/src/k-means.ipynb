### k-means 클러스터링
##### 참고 자료 : https://velog.io/@73syjs/K-means-clustering


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs  # 클러스터링 예제 데이터 생성을 위한 함수
from sklearn.cluster import KMeans  # K-평균 클러스터링을 위한 함수
```

```python
# 무작위 데이터 생성
locations, labels = make_blobs(n_samples=100, centers=3, n_features=2)  
# 총 100개의 샘플, 3개의 클러스터 중심, 2개의 특성을 가진 데이터를 생성합니다.
```

```python
print(locations)  # 생성된 무작위 위치 데이터를 출력합니다.
```

```python
print(labels)  # 각 데이터 포인트의 클러스터 레이블을 출력합니다.
```

```python
# 데이터프레임 생성
df = pd.DataFrame(dict(x=locations[:, 0], y=locations[:, 1], label=labels))

# 데이터프레임 출력
print(df)  # 데이터프레임을 출력하여 x, y 좌표 및 레이블 정보를 확인합니다.
```

```python
# 생성된 데이터 시각화
dx = df.iloc[:, 0].values  # x 좌표 값 추출
dy = df.iloc[:, 1].values  # y 좌표 값 추출

plt.scatter(dx, dy)  # 산점도를 사용하여 데이터 포인트를 시각화합니다.
```

```python
# k-means 클러스터링
kmeans = KMeans(n_clusters=3)  # k=3으로 KMeans 클러스터링 객체 생성
kmeans.fit(locations)  # 데이터에 KMeans 클러스터링 적용
pred_labels = kmeans.predict(locations)  # 예측된 클러스터 레이블을 가져옵니다.

print('예측된 레이블 = \n', pred_labels)  # 예측된 클러스터 레이블을 출력합니다.
```

```python
# 원래의 레이블을 사용한 데이터 시각화
colors = np.array([0, 10, 20])  # 색상 배열 정의

plt.scatter(dx, dy, c=colors[labels])  # 각 데이터 포인트에 원래의 레이블에 해당하는 색상을 적용하여 시각화합니다.
```

```python
# 클러스터링 결과를 사용한 데이터 시각화
colors = np.array([0, 10, 20])  # 색상 배열 정의

plt.scatter(dx, dy, c=colors[pred_labels])  # 각 데이터 포인트에 클러스터링 결과에 따른 색상을 적용하여 시각화합니다.
```

```python
# KMeans 클러스터링의 SSE(Sum of Squared Errors) 계산
sum_of_squared_distances = []  # SSE 값을 저장할 리스트
K = range(1, 15)  # 클러스터 개수 K의 범위

# K 값에 따른 SSE 계산
for k in K:
    km = KMeans(n_clusters=k)  # KMeans 클러스터링 객체 생성
    km = km.fit(df)  # 데이터에 클러스터링 적용
    sum_of_squared_distances.append(km.inertia_)  # 해당 K 값에 대한 SSE 값을 리스트에 추가

# 엘보우 메서드를 통한 최적 K 선택을 위한 시각화
plt.plot(K, sum_of_squared_distances, 'ro-')  # K 값에 따른 SSE 그래프
plt.xlabel('k')  # x축 레이블 설정
plt.ylabel('Sum_of_squared_distances')  # y축 레이블 설정
plt.title('Elbow Method For Optimal k')  # 그래프 제목 설정
plt.show()  # 그래프 출력
```
